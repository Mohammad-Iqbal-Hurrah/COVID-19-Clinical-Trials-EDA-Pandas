```python
# -----------------------------------------------------------
# Project: COVID-19 Clinical Trials EDA using Pandas
# Objective: Explore patterns and insights from global clinical trial data
# Dataset Source: ClinicalTrials.gov (CSV extract)
# Author: Mohammad Iqbal Hurrah
# -----------------------------------------------------------

# -----------------------------------------------------------
# 1. Import Required Libraries
# These libraries support data handling, numerical ops & visualization.
# -----------------------------------------------------------
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# -----------------------------------------------------------
# 2. Load dataset
# index_col=0 means first column treated as row index.
# -----------------------------------------------------------
df = pd.read_csv("COVID clinical trials.csv", index_col=0)

# -----------------------------------------------------------
# 3. Initial quick view of dataset
# Helps understand fields & values before cleaning.
# -----------------------------------------------------------
df.head()
df.shape          # total rows & columns
df.columns        # column names

# -----------------------------------------------------------
# 4. Identify column data types
# Helps to decide missing-value strategy per column type.
# -----------------------------------------------------------
categorical_cols = df.select_dtypes(include="object").columns
numeric_cols     = df.select_dtypes(exclude="object").columns

# -----------------------------------------------------------
# 5. Detect missing values (percentage)
# Useful to decide drop/impute strategies.
# -----------------------------------------------------------
missing_percent = df.isnull().mean() * 100
missing_percent

# -----------------------------------------------------------
# Helper function to visualize distribution bars
# Displays top 40 entries to avoid clutter.
# -----------------------------------------------------------
def visualize_data(series, title="", ylabel="Count / %"):
    sns.set(rc={'figure.figsize':(15,8)})
    plt.xticks(rotation=90)
    sns.barplot(x=series.index[:40], y=series.values[:40])
    plt.title(title)
    plt.ylabel(ylabel)
    plt.show()

# visualize missing %
visualize_data(missing_percent, title="Missing Value Percentage by Feature")

# -----------------------------------------------------------
# 6. Drop columns with >95% missing values
# Justification: too sparse to generate meaningful signals.
# These disturb analysis and inflate cleaning overhead.
# -----------------------------------------------------------
# df.drop(['Results First Posted', 'Study Documents'], axis=1, inplace=True)

# -----------------------------------------------------------
# 7. Drop duplicate records if any
# -----------------------------------------------------------
df.drop_duplicates(inplace=True)

# -----------------------------------------------------------
# 8. Remove rows having less than 10 non-null fields
# Ensures each row contains enough analytical context.
# -----------------------------------------------------------
df.dropna(thresh=10, inplace=True)

# -----------------------------------------------------------
# 9. Extract country information from Locations column
# Locations field contains full address; country exists at end.
# Feature engineering for geographic analysis.
# -----------------------------------------------------------
df['Country'] = df['Locations'].apply(lambda x: str(x).split(',')[-1])

# -----------------------------------------------------------
# 10. Handle categorical missing values
# Strategy: create a distinct "Missing X" category.
# Rationale: categorical NA often carries semantic meaning.
# -----------------------------------------------------------
for col in categorical_cols:
    df[col] = df[col].fillna(f"Missing {col}")

# -----------------------------------------------------------
# 11. Handle missing numeric values (Enrollment feature)
# Observed strong right-skew → median imputation avoids bias.
# -----------------------------------------------------------
df['Enrollment'] = df['Enrollment'].fillna(df['Enrollment'].median())

# -----------------------------------------------------------
# 12. Visualization – Key EDA findings
# -----------------------------------------------------------

# Top 10 countries by no. of trials
top_countries = df['Country'].value_counts().head(10)
visualize_data(top_countries, "Top Trial Conducting Countries", "Trial Count")

# Application status distribution
status_dist = df['Status'].value_counts()
visualize_data(status_dist, "Trial Status Distribution")

# Gender distribution
gender_dist = df['Gender'].value_counts()
visualize_data(gender_dist, "Gender Distribution")

# Time trend – extract month from start date string
start_month = df['Start Date'].apply(lambda x: str(x).split(" ")[0])
month_dist = start_month.value_counts()
visualize_data(month_dist, "Trials Started Per Month")

# -----------------------------------------------------------
# 13. Suggested ML preparation indicator (not executed here)
# Unique IDs not useful as predictive features.
# If ML planned, drop them.
# -----------------------------------------------------------
print("Unique NCT IDs:", df['NCT Number'].nunique())
print("Unique URL values:", df['URL'].nunique())

# -----------------------------------------------------------
# End of EDA Notebook
# -----------------------------------------------------------
```
