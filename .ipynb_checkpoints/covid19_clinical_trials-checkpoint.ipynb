```python
# IMPORT NECESSARY LIBERARIES 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px        #WE WILL DRAW WORLD MAP USING THIS 
import seaborn as sns
```

# DATA EXPLORATION

```python
#Reading data set 
df = pd.read_csv("COVID clinical trials.csv")

# Quick structural overview
print("\nDATASET SUMMARY\n")
print(df.info())              # schema + null counts + dtype
print("\nNumeric Summary\n")
print(df.describe())
print("\nCategorical Summary\n")
print(df.describe(include='object'))

#Review Initial Data
print("\nFirst 5 rows:\n") 
print(df.head())              #Show first 5 rows of the dataset
print("\nLast 5 rows:\n")
print(df.tail())              #Show last 5 rows of the dataset
```

# HANDLING MISSING VALUES AND FEATURE ENGINEERING.

```python
# For selected categorical columns where nulls are meaningful

missing_percentage = df.isnull().mean()*100 #calculate missing value percentage
print(missing_percentage) #All columns with missing percentage 

cols_to_drop = missing_percentage[missing_percentage>70].index    #take columns with missing percentage rate >70
df.drop(columns = cols_to_drop,inplace=True)             #drop those columns 
# ['Results First Posted', 'Study Documents'] Dropped

categorical_to_impute = ['Acronym','Interventions','Phases','Locations']    #Categorical columns having missing/null values
df[categorical_to_impute] = df[categorical_to_impute].fillna("Missing Category") #filling nulls

# Enrollment is numeric + skewed heavily → median suits imputation
enrollment_median = df['Enrollment'].median() #extract enrollment column median
df['Enrollment'] = df['Enrollment'].fillna(enrollment_median) #fill missing values with median value 

# Remaining sparse nulls replaced uniformly for consistency
df.fillna("MISSING", inplace=True)

print("\nAfter cleaning, missing count per column:")
print(df.isnull().sum()) #Data Cleaned
```

# VISUALIZATION

# Clinical Trial Status Frequency

```python


# Trial status distribution (UNIVARIATE)
status_dist = df['Status'].value_counts()
print(status_dist)

plt.figure(figsize=(12,6))
sns.barplot(x=status_dist.index, y=status_dist.values)
plt.title("Clinical Trial Status Frequency")
plt.xlabel("Status")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()

# Insights: 
# The distribution shows that 'Recruiting' trials are dominant.
# which reflects the late-pandemic stabilization period.
# that follow-ups & regulatory steps are ongoing.
```

# Distribution of Clinical Trial Phases

```python
# Phase distribution (UNIVARIATE)
phase_dist = df['Phases'].value_counts()
print(phase_dist)

plt.figure(figsize=(10,6))
sns.barplot(x=phase_dist.index, y=phase_dist.values)
plt.title('Distribution of Clinical Trial Phases')
plt.xlabel('Trial Phase')
plt.ylabel('Number of Trials')
plt.xticks(rotation=90)
plt.show()
```

# Gender Participation in Trials

```python
# Gender inclusion distribution (UNIVARIATE)
gender_dist = df['Gender'].value_counts()
print(gender_dist)

plt.figure(figsize=(12,6))
sns.barplot(x=gender_dist.index, y=gender_dist.values)
plt.title("Gender Participation in Trials")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Insight for Gender Participation in Trials.
# Majority of clinical trials include all genders,
# confirming balanced participant eligibility.
# Very small portion of women-only/male-only trials suggests 
# that COVID-19 research remained largely generalized
# rather than gender-specific.
```

# Top 10 Countries in Clinical Trials

```python
#FEATURE ENGINEERING 
# Extract country column 
df['Country'] = df['Locations'].astype(str).apply(lambda row: row.split(',')[-1].strip()) #extract last part  from location column 

#Top 10 countries
top_countries = df['Country'].value_counts().head(10) 
print(top_countries)

plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries.values)
plt.title("Top 10 Countries in Clinical Trials")
plt.xlabel("Country")
plt.ylabel("Trial Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Insight for Top 10 Countries in Clinical Trials:
# United States has the largest number of trials, reflecting
# its medical research infrastructure and funding capacity.
# European countries follow, indicating global coordination
# toward COVID-19 responses during the pandemic peak.
```

# Enrollment Distribution Across Trials

```python
plt.figure(figsize=(10,6))
sns.histplot(df['Enrollment'], bins=50, kde=True)
plt.title("Enrollment Distribution Across Trials")
plt.xlabel("Enrollment Count")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()

# Insight Enrollment Distribution Across Trials:
# Enrollment distribution is right-skewed,
# meaning few trials recruited very large participants
# while most studies ran with smaller groups.
# Median imputation earlier was justified.
```

# Trials Status Across Phases

```python
# Make a cross tab first
status_phase = pd.crosstab(df['Status'], df['Phases'])

# just to check the values 
print("Status vs. Phases Table:")
# print(status_phase)

# Status vs. Phases Distribution
status_phase.plot(kind='bar', stacked=True, figsize=(12, 7))
plt.title('Status vs. Phases Distribution')
plt.xlabel('Trial Status')
plt.ylabel('Number of Trials')
plt.legend(title='Phases', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=90)
plt.show()

# Insight for Trials Status Across Phases:
# Phase 3 and Phase 4 trials show higher completion rates,
# as late-stage trials are typically more structured & well-funded.
# Early phase trials have more withdrawn or unknown statuses
# indicating research uncertainty during early pandemic.
```

# Conditions vs. Outcome Measures

```python
# fill missing values 
df['Outcome Measures'] = df['Outcome Measures'].fillna('Missing')

# Conditions vs Outcome Measures
conditions_outcomes = df.groupby('Conditions')['Outcome Measures'].apply(lambda x: ', '.join(x)).reset_index()

print("Top 5 Conditions and their Outcome Measures:")
print(conditions_outcomes.head())
```

# Trend of Trials Initiated by Month

```python
# convert date column into datetime format -->coerce means ignore erros for wrong formats
df['Start Date'] = pd.to_datetime(df['Start Date'], errors='coerce')
df['Primary Completion Date'] = pd.to_datetime(df['Primary Completion Date'], errors='coerce')

# monthly trends 
plt.figure(figsize=(12, 6))
df['Start Date'].dt.to_period('M').value_counts().sort_index().plot(kind='line', marker='o', color='red')

plt.title('COVID-19 Trials Started Over Time')
plt.xlabel('Month-Year')
plt.ylabel('Number of Trials')
plt.grid(True)
plt.show()

# Insight for Trend of Trials Initiated by Month :
# March–May months show peak in new trial start counts
# This period aligns with initial pandemic escalation.
# prompting accelerated clinical research responses worldwide.
```

# Correlation Heatmap

```python
# Correlation Heatmap (Only for Numeric Columns)
plt.figure(figsize=(10, 6))

# numeric_only=True likhna zaroori hai errorse bachne k liye
correlation_matrix = df.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Numeric Features")
plt.show()

# Insight for Correlation Heatmap of Numeric Features:\nThis heatmap shows the relationship between different numeric columns.
# It helps us check if 'Enrollment' numbers are related to the 'Rank' of the trial.
# Most values are near zero, which means there is no strong link between these features.
```

# Interactive World Map

```python
country_counts = df['Country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Trial Count']

# Interactive World Map
fig = px.choropleth(country_counts, 
                    locations="Country", 
                    locationmode='country names',
                    color="Trial Count", 
                    hover_name="Country", 
                    title="Global Distribution of COVID-19 Clinical Trials",
                    color_continuous_scale=px.colors.sequential.Plasma)

fig.show()

# Insight: 
# This interactive map shows how COVID trials are spread across the world.
# We can see that North America and Europe have the highest number of trials.
#Just hover over map and see the magic
```
